```python
df=df.drop('row ID',axis=1)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute  import SimpleImputer
imputer=SimpleImputer(strategy='median')
df['MinTemp']=imputer.fit_transform(df[['MinTemp']])
df['MaxTemp']=imputer.fit_transform(df[['MaxTemp']])
df['Rainfall']=imputer.fit_transform(df[['Rainfall']])
df['Evaporation']=imputer.fit_transform(df[['Evaporation']])
df['Sunshine']=imputer.fit_transform(df[['Sunshine']])
df['WindGustSpeed']=imputer.fit_transform(df[['WindGustSpeed']])
df['WindSpeed9am']=imputer.fit_transform(df[['WindSpeed9am']])
df['WindSpeed3pm']=imputer.fit_transform(df[['WindSpeed3pm']])
df['Humidity9am']=imputer.fit_transform(df[['Humidity9am']])
df['Humidity3pm']=df['Humidity3pm'].fillna(df['Humidity3pm'].median())
df['Pressure9am']=df['Pressure9am'].fillna(df['Pressure9am'].median())
df['Pressure3pm']=df['Pressure3pm'].fillna(df['Pressure3pm'].median())
df['Cloud9am']=df['Cloud9am'].fillna(df['Cloud9am'].median())
df['WindGustDir']=df['WindGustDir'].fillna(df['WindGustDir'].mode()[0])
df['WindDir9am']=df['WindDir9am'].fillna(df['WindDir9am'].mode()[0])
df['WindDir3pm']=df['WindDir3pm'].fillna(df['WindDir3pm'].mode()[0])
df['RainToday']=df['RainToday'].fillna(df['RainToday'].mode()[0])
```

```python
imputer=SimpleImputer(strategy='mean')
df['Cloud3pm']=imputer.fit_transform(df[['Cloud3pm']])
df['Temp9am']=imputer.fit_transform(df[['Temp9am']])
df['Temp3pm']=imputer.fit_transform(df[['Temp3pm']])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
le=LabelEncoder()
oe=OrdinalEncoder()
df['Location']=le.fit_transform(df['Location'])
df['WindGustDir']=le.fit_transform(df['WindGustDir'])
df['WindDir9am']=le.fit_transform(df['WindDir9am'])
df['WindDir3pm']=le.fit_transform(df['WindDir3pm'])
df['RainToday']=oe.fit_transform(df[['RainToday']])
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x=ms.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
y_pred=sv.predict(x_test)
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
