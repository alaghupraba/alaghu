```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.boxplot(x=df['FUEL CONSUMPTION'])
plt.show()
```

```python
Q1=df['FUEL CONSUMPTION'].quantile(0.25)
Q3=df['FUEL CONSUMPTION'].quantile(0.75)
IQR=Q3-Q1
print(IQR)
lower_bound=Q1-1.5*IQR
upper_bound=Q1+1.5*IQR
print(lower_bound)
print(upper_bound)
outliers=df[(df['FUEL CONSUMPTION']<lower_bound)|(df['FUEL CONSUMPTION']>upper_bound)]
print(outliers)
df=df[~((df['FUEL CONSUMPTION']>=(Q1-1.5*IQR))&(df['FUEL CONSUMPTION']<=(Q3+1.5*IQR)))]
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['FUEL CONSUMPTION'])
plt.show()
```

```python
x=df.drop('FUEL CONSUMPTION',axis=1)
y=df['FUEL CONSUMPTION']
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
